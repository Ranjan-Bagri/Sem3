```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def z1(x,y):
    return (1-(x**2+y**3))*np.exp(-(x**2+y**2)/2)

x=y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
z=z1(x,y)
fig=plt.figure(figsize=(5,4))
ax=plt.axes(projection="3d")
surf=ax.plot_surface(x,y,z,cmap='Reds')
plt.show()
```

```python
def z2(x,y):
    return np.sin(np.pi*x/3)*np.sin(np.pi*y/3)
z=z2(x,y)
fig=plt.figure(figsize=(5,4))
ax=plt.gca(projection="3d")
surf=ax.plot_surface(x,y,z,cmap='Blues')
plt.show()
```

```python
def z3(x,y):
    return np.sin(x/1.5)**10+np.cos(10+y*x/1.5)*np.cos(x/1.5)
z=z3(x,y)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
surf=ax.plot_surface(x,y,z,cmap='Greens')
plt.show()
```

**Intensity of Fraunhofer Driffraction for rectangular aperture**

$I_R(x,y)=I_0\left(\frac{\sin(sax)}{sax}\right)^2 \left(\frac{\sin(say)}{say}\right)^2 , s= \frac{\pi}{\lambda R}=1$

**Intensity of Fraunhofer Driffraction for circular aperture**

$I_C(x,y)=I_0\left(\frac{J_1(2sa\sqrt{x^2+y^2})}{2sa\sqrt{x^2+y^2}}\right)^2 , s= 1$

```python
from scipy.special import j1

def intsRect(x,y):
    a=4
    b=2*a
    al,be=a*x,b*y
    return (np.sin(al)/al*np.sin(be)/be)**2

def intsCirc(x,y):
    a=4
    q=(x**2+y**2)**0.5
    I=j1(2*a*q)/(2*a*q)
    return I

x=y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
fig=plt.figure(figsize=(9,4))

# plot the intensity due to rectangular aperture
fig.add_subplot(121,projection="3d")
z=intsRect(x,y)
surf=ax.plot_surface(x,y,z,cmap='RdBu')
plt.title(r'$I_R$')
fig.colorbar(surf,shrink=0.5)

# plot of intensity due to circular aperture
fig.add_subplot(122,projection='3d')
z=intsCirc(x,y)
surf=ax.plot_surface(x,y,z,cmap='pink')
plt.title(r'$I_C$')
fig.colorbar(surf,shrink=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import proj3d

def z1(x,y):
    return np.sin(np.pi*x)*np.sin(np.pi*y)

x=np.linspace(-1,1,50)
y=np.linspace(-1,1,50)
x,y=np.meshgrid(x,y)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
z=z1(x,y)
ax.plot_surface(x,y,z,cmap='afmhot')
ax.contour(x,y,z,100,offset=-1,cmap="afmhot")
plt.show()
```

```python
# Plot 3d polar curve
R=1.0
#meshgrid for 3d plots
theta,phi=np.linspace(0.0,np.pi,100),np.linspace(0.0,2*np.pi,100)
theta,phi=np.meshgrid(theta,phi)

# coordinates for sphere
x=R*np.sin(theta)*np.cos(phi)
y=R*np.sin(theta)*np.sin(phi)
z=R*np.cos(theta)

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
# plot of the sphere
ax.plot_surface(x,y,z,color='peachpuff')
#ax.set_aspect('equal')
plt.show()
```

```python
# saving a plot
R,a=2.0,1.0
t=np.arange(-10,10,0.01)
x,y,z=R*np.cos(t),R*np.sin(t),a*t
fig=plt.figure()
ax=plt.gca(projection='3d')
ax.plot(x,y,z,color='k')
plt.title('Helix')
plt.savefig('helix.png')
plt.show()
```
