```python
import numpy as np
from matplotlib import pyplot as plt
def f(x):
    return x**2*np.exp(-x)
x=np.linspace(0,10,50)
plt.plot(x,f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
def f(mu,sig,x):
    return (1/(2*np.pi*sig**2)**0.5)*np.exp(-(x-mu)**2/(2*sig**2))
x=np.arange(-5.0,5.0,0.01)
fx1=f(0,0.5,x)
fx2=f(0,1.2,x)
fx3=f(-2,0.7,x)
plt.plot(x,fx1,x,fx2,x,fx3)
plt.title(r'$f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
x=np.arange(0,2.0,0.01)
plt.suptitle('Multiple plots or subplots')

fig=plt.figure()
fx1=fig.add_subplot(2,2,2)
fx2=fig.add_subplot(2,2,4)
fx3=fig.add_subplot(2,2,(1,3))

fx1.plot(x,x*np.exp(-x),label=r'$xe^{-x}$')
fx1.legend(loc='best',prop={'size':12})

fx2.plot(x,x**2*np.exp(-2*x),label=r'$x^2e^{-2x}$')
fx2.legend(loc='best',prop={'size':12})
plt.xlabel(x)

fx3.plot(x,x**3*np.exp(-5*x),label=r'$x^3e^{-5x}$')
fx3.legend(loc='best',prop={'size':12})
plt.xlabel(x)

plt.show()
```

```python
def f(x):
    return 1/(1+x**2)+0.1/(1+(10*(3-x))**2)

fig=plt.figure(figsize=(8,6))
plt.suptitle(r'subplots for f(x) = $\frac{1}{1+x^2}+\frac{0.1}{1+[10(3-x)]^2}$')
x0, x1=-4, 14
ax=fig.add_axes([0.1,0.15,0.8,0.8])
x=np.linspace(x0,x1,1000)
ax.plot(x,f(x),label=r'$-4\leq x \leq 14$')
ax.legend(loc='lower right',prop={'size':12})
plt.xlabel("x")

x0,x1=2.5,3.5
ax=fig.add_axes([0.5,0.5,0.38,0.42])
x=np.linspace(x0,x1,1000)
ax.plot(x,f(x),label=r'$2.5\leq x \leq 3.5$')
ax.legend(loc='best',prop={'size':12})
plt.xlabel("x")
plt.show()
```

```python
def f(a,x):
    return np.sin(a*x)
x=np.linspace(-np.pi,np.pi,40)
a=-1.0
while a<=1.0:
    plt.plot(x,f(a,x),label=r'sin(%.2fx)'%a)
    plt.xlabel('x')
    plt.legend(loc='best',prop={'size':12})
    plt.show()
    a+=0.5
```

```python
# creation of lissajous figures
def f1(th):
    return np.sin(th)
def f2(a,n,th,dl):
    return a*np.sin(n*th+dl)
a,n,dl=1,2,0

while dl<=2*np.pi:
    th=np.arange(-2*np.pi,2*np.pi,0.01)
    x=f1(th)
    y=f2(a,n,th,dl)
    plt.plot(x,y)
    plt.xlabel(r'$\sin \theta$')
    plt.ylabel(r'$\sin(2\theta+\delta)$')
    plt.pause(1.0)
    dl+=np.pi/5
plt.show()
```

$x(t)=u\cos{\theta t}$

$y(t)=u\sin{\theta t}-\frac{1}{2}gt^2$

```python
# This is a demonstration of real time plotting of Projectile motion
g,u,theta=9.8,8.0,np.pi/4
t=0.0
x,y=[],[]


while True:
    xx=u*np.cos(theta)*t
    yy=u*np.sin(theta)*t-0.5*g*t**2
    if yy<0.0:
        break
    x.append(xx)
    y.append(yy)
    plt.clf()
    plt.plot(x,y,'k.')
    plt.xlabel("x(t)")
    plt.ylabel("y(t)")
    plt.title('Projectile motion')
    plt.pause(0.05)
    t+=0.01
plt.show()
```

```python
# Demonstration of different line properties
color=['k','b','r','g']
linestyle=['-','--','-.',':']
linewidth=[1,2.1,5.2,6.3]
x=np.arange(0,2.2,0.05)

plt.subplot(1,3,1)
for i in range(4):
    plt.plot(x,x*np.exp(-(i+1)*x),linestyle[i],lw=2,c='k',label='ls='+"'"+linestyle[i]+"'")
plt.legend(loc='best',prop={'size':12})
plt.title('Demonstration of line style')
plt.xlabel(x)

plt.subplot(1,3,2)
for i in range(4):
    plt.plot(x,x**2*np.exp(-(i+1)*x),ls='-',lw=linewidth[i],c='k',label='lw='+'%.1f'%linewidth[i])
plt.legend(loc='best',prop={'size':12})
plt.title('Demonstration of line width')
plt.xlabel(x)

plt.subplot(1,3,3)
for i in range(4):
    plt.plot(x,x**3*np.exp(-(i+1)*x),ls='-',lw=2,c=color[i],label='c='+"'"+color[i]+"'")
plt.legend(loc='best',prop={'size':12})
plt.title('Demonstration of line width')
plt.xlabel(x)

plt.show()
```

```python
# Demonstration of marker properties
def crc(th):
    return 4*np.cos(th),4*np.sin(th)
marker_types=['.','*','o','v','^','<','>','1','2','3','+','x']
cols=['b','r','k','g']
marker_size=[2.2,3.2,4,5.1]

plt.axes().set_aspect('equal')
th=np.linspace(0,2*np.pi,len(marker_size),endpoint=False)
x,y=crc(th)
plt.subplot(1,2,1)
for i in range(len(x)):
    plt.plot(x[i],y[i],c='k',marker=marker_size[i],ms=6)
    plt.text(x[i]+0.1,y[i]+0.1,marker_size[i])

plt.subplot(1,2,2)
x=np.arange(0,2.2,0.05)
for i in range(len(marker_size)):
    plt.plot(x,(x+2)*np.exp(-(i+1)*x),'o',c=cols[i],ms=marker_size[i],label='ms=%d'%marker_size[i]+r'$(x+2)e^{-%d x}$'%(i+1))

plt.show()
```
