```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Demonstration of plot with line and marker
def f(pr,x):
    return np.sin(x)*pr
lnmarsty=['bo-','r^--','gD:']
x=np.linspace(-np.pi,np.pi,30)
for i in range(len(lnmarsty)):
    plt.plot(x,f(i+1,x),lnmarsty[i])
plt.show()
```

```python
# Displaying texts
def f(x):
    return np.tan(x)
x=np.arange(-np.pi,np.pi,0.01)
y=f(x)
plt.plot(x,y)
plt.ylim(-100,100)
plt.text(-1,65,'tan(x)',fontsize=14,family='serif',color='g',rotation=15)
plt.show()
```

### Different kinds of plots

```python
def f(x):
    return x**1.5*np.exp(-x)
x=np.arange(1e-3,10,0.5e-2)

plt.suptitle(r'$f(x)=x^{3/2}e^-{-x}$')

plt.subplot(2,2,1)
plt.plot(x,f(x),'k')
plt.text(5.0,0.3,'normal plot')

plt.subplot(2,2,2)
plt.semilogx(x,f(x),'k')
plt.text(5e-3,0.3,'semilogx plot')

plt.subplot(2,2,3)
plt.semilogy(x,f(x),'k')
plt.text(3.0,1e-3,'semilogy plot')

plt.subplot(2,2,4)
plt.loglog(x,f(x),'k')
plt.text(5e-2,5e-4,'loglog plot')

plt.show()
```

$r(\theta)=\frac{\theta}{2\pi}$

```python
# Demonstration of polar plot
def r(th):
    return 1/(2*np.pi)*th
th=np.arange(0,4*np.pi,0.01)
plt.polar(th,r(th),'k')
plt.title('Single Polar plot')
plt.show()
```

$r(\theta)=\frac{de}{1+\cos{\theta}}$

```python
def r(d,e,th):
    return (d*e)/(1+e*np.cos(th))
th=np.arange(0,2*np.pi,0.01)
fig=plt.figure(figsize=(10,12))

# circle
ax=fig.add_subplot(231,polar=True)
ax.plot(th,r(1.5,0,th))
ax.set_rmax(0.02)
ax.legend(['e=0'])

# parabola
ax=fig.add_subplot(232,polar=True)
ax.plot(th,r(1.5,1,th))
ax.set_rmin(0)
ax.set_rmax(30)
ax.legend(['e=1'])

# ellipse
ax=fig.add_subplot(233,polar=True)
ax.plot(th,r(1.5,0.5,th))
ax.set_rmin(0)
ax.set_rmax(2.02)
ax.legend(['e=0.5'])

# hyperbola 1
ax=fig.add_subplot(234,polar=True)
ax.plot(th,r(1.5,2,th))
ax.set_thetamin(100)
ax.set_thetamax(150)
ax.set_rmin(-100)
ax.set_rmax(100)
ax.legend(['e=2'])

# hyperbola 2
ax=fig.add_subplot(235,polar=True)
ax.plot(th,r(1.5,2,th))
ax.set_thetamin(210)
ax.set_thetamax(260)
ax.set_rmin(-100)
ax.set_rmax(100)
ax.legend(['e=2'])

plt.show()
```

```python
# parametric plot
a=2.0
t=np.arange(0,2*np.pi,0.01)
x=a*np.cos(t)
y=a*np.sin(t)
plt.plot(x,y,lw=2,label=r'$x(t)= \cos(t), y(t)= \sin(t)$')
plt.axis('equal')
plt.title('Parametric Plot')
plt.legend(loc='best',prop={'size':12})
plt.show()
```

```python
# python function to visualize fringes
def fringes(x1,x2,y1,y2,ints):
    # Arguments
    # x1,x2 ==> x-range of the screen
    # y1,y2 ==> y-range of the screen
    # ints  ==> intensity function ints(x,y)
    # Reurns
    # fig ==> matplotlib figure object
    # ax0 ==> matplotlib axes object
    dx,dy = 0.01,0.01
    x,y=np.meshgrid(np.arange(x1,x2,dx),np.arange(y1,y2,dy))
    np.seterr(divide='ignore',invalid='ignore')
    z=ints(x,y)
    fig=plt.figure()
    ax0=fig.add_subplot(111)
    im=ax0.pcolormesh(x,y,z,cmap='gray')
    fig.colorbar(im,ax=ax0)
    plt.title('fringes')
    plt.xlabel('x')
    plt.ylabel('y')
    return fig,ax0
```

$I=I_0 \cos^2\left(\frac{\phi}{2}\right)$ , $\phi=\frac{2\pi yd}{\lambda D}$

```python
def ints(x,y):
    d=0.05
    lamdaD=0.06 # lambda * D = 0.06
    phi=2*np.pi*y*d/lamdaD
    return np.cos(phi/2)**2

x1,x2,y1,y2=-5.05,5.05,-5.05,5.05
fig,ax0=fringes(x1,x2,y1,y2,ints)
#plt.title("Fringes of Young's double slit experiment")
plt.show()

    
```

```python
# Hadinger fringes
def ints(x,y):
    dlmb=30# d/lmb
    mu1,mu2 = 1.0,1.5
    D=2.0
    phi=4*np.pi*dlmb*(mu2**2-(mu1**2*(x**2+y**2))/(x**2+y**2+D**2))+np.pi
    return np.cos(phi/2)**2

x1,x2,y1,y2= -0.5,0.5,-0.5,0.5
fig,ax = fringes(x1,x2,y1,y2,ints)
plt.title('Hadinger fringes')
plt.show()
```

```python
def ints(x,y):
    lmbR=0.5
    mu=1.5
    phi=(2*np.pi*mu)/lmbR*(x**2+y**2)+np.pi
    return np.cos(phi/2)**2
x1,x2,y1,y2=-1.5,1.5,-1.5,1.5
fig,ax=fringes(x1,x2,y1,y2,ints)
plt.title("Newton's Ring")
plt.show()
```
