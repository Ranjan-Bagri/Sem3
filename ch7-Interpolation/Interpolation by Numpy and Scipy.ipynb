**Interpolation with Numpy**

```python
import numpy as np
```

```python
x=[1,2,3] # array of x-values
fx=[3,2,0] # array of y-values
xx=[0,1,1.5,2.72,3.14] # x-values where interpolations are required
```

```python
np.interp(xx,x,fx) # linear interplation
```

**Interpolation with Scipy**

Syntax: `scipy.interpolate.interp1d(x,fx,kind)`

```python
from scipy.interpolate import interp1d
```

```python
x=np.linspace(0,10,num=11,endpoint=True)
x
```

```python
y=np.cos(-x**2/9.0) # create y-values from a function
y
```

```python
f=interp1d(x,y) # function object corresponding to linear interpolation
f
```

```python
f2=interp1d(x,y,kind='cubic') # function object for linear interplation
f2
```

```python
xnew=np.linspace(0,10,num=41,endpoint=True)
xnew
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew),'--')
plt.legend(['data','linear','cubic'],loc='best') # legend object
plt.show()
```
